```python
#BTM1
import math
import copy

X = "X"
O = "O"
EMPTY = " "

def initial_state():
    return [[EMPTY]*3 for _ in range(3)]

def print_board(board):
    for row in board:
        print("|".join(row))
        print("-"*5)

def player(board):
    count = sum(row.count(X) + row.count(O) for row in board)
    return X if count % 2 == 0 else O

def actions(board):
    return [(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY]

def result(board, action):
    new_board = copy.deepcopy(board)
    i, j = action
    new_board[i][j] = player(board)
    return new_board

def winner(board):
    lines = []
    lines.extend(board)
    lines.extend([[board[i][j] for i in range(3)] for j in range(3)])
    lines.append([board[i][i] for i in range(3)])
    lines.append([board[i][2-i] for i in range(3)])
    for line in lines:
        if line[0] != EMPTY and line.count(line[0]) == 3:
            return line[0]
    return None

def terminal(board):
    return winner(board) is not None or all(cell != EMPTY for row in board for cell in row)

def utility(board):
    w = winner(board)
    if w == X: return 1
    if w == O: return -1
    return 0

def max_value(board):
    if terminal(board): return utility(board)
    v = -math.inf
    for action in actions(board):
        v = max(v, min_value(result(board, action)))
    return v

def min_value(board):
    if terminal(board): return utility(board)
    v = math.inf
    for action in actions(board):
        v = min(v, max_value(result(board, action)))
    return v

def minimax(board):
    if player(board) == X:
        best_val = -math.inf
        best_move = None
        for action in actions(board):
            val = min_value(result(board, action))
            if val > best_val:
                best_val = val
                best_move = action
    else:
        best_val = math.inf
        best_move = None
        for action in actions(board):
            val = max_value(result(board, action))
            if val < best_val:
                best_val = val
                best_move = action
    return best_move

def play():
    board = initial_state()
    # Người chơi sẽ là X (đi trước), AI sẽ là O (đi sau)
    while not terminal(board):
        print_board(board)
        if player(board) == X:
            try:
                row, col = map(int, input("Lượt của bạn (X) - Nhập hàng (1-3) cột (1-3): ").split())
                move = (row - 1, col - 1)
                if move not in actions(board):
                    print("Nước đi không hợp lệ!")
                    continue
            except:
                print("Vui lòng nhập 2 số cách nhau bởi khoảng trắng.")
                continue
        else:
            print("AI (O) đang tính toán...")
            move = minimax(board)
            print("AI chọn:", (move[0] + 1, move[1] + 1))
        board = result(board, move)

    print_board(board)
    res = winner(board)
    if res: print(f"Người thắng là: {res}")
    else: print("Hòa!")

play()
```

```python
#BTM2
import math
import copy

X = "X"
O = "O"
EMPTY = " "

def initial_state():
    return [[EMPTY]*3 for _ in range(3)]

def print_board(board):
    for row in board:
        print("|".join(row))
        print("-"*5)

def player(board):
    count = sum(row.count(X) + row.count(O) for row in board)
    return X if count % 2 == 0 else O

def actions(board):
    return [(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY]

def result(board, action):
    new = copy.deepcopy(board)
    i, j = action
    new[i][j] = player(board)
    return new

def winner(board):
    lines = []
    lines.extend(board)
    lines.extend([[board[i][j] for i in range(3)] for j in range(3)])
    lines.append([board[i][i] for i in range(3)])
    lines.append([board[i][2-i] for i in range(3)])

    for line in lines:
        if line[0] != EMPTY and line.count(line[0]) == 3:
            return line[0]
    return None

def terminal(board):
    return winner(board) or all(cell != EMPTY for row in board for cell in row)

def utility(board):
    w = winner(board)
    if w == X: return 1
    if w == O: return -1
    return 0

# Alpha-beta
def alphabeta(board, alpha, beta, maximizing):
    if terminal(board):
        return utility(board)

    if maximizing:
        value = -math.inf
        for action in actions(board):
            value = max(value, alphabeta(result(board, action), alpha, beta, False))
            alpha = max(alpha, value)
            if alpha >= beta:
                break
        return value
    else:
        value = math.inf
        for action in actions(board):
            value = min(value, alphabeta(result(board, action), alpha, beta, True))
            beta = min(beta, value)
            if beta <= alpha:
                break
        return value

def best_move(board):
    best = None
    if player(board) == X:
        best_val = -math.inf
        for action in actions(board):
            val = alphabeta(result(board, action), -math.inf, math.inf, False)
            if val > best_val:
                best_val = val
                best = action
    else:
        best_val = math.inf
        for action in actions(board):
            val = alphabeta(result(board, action), -math.inf, math.inf, True)
            if val < best_val:
                best_val = val
                best = action
    return best

def play():
    board = initial_state()
    while not terminal(board):
        print_board(board)
        if player(board) == X:
            move = best_move(board)
            print("AI chọn:", move)
        else:
            move = tuple(map(int, input("Nhập: ").split()))
        board = result(board, move)

    print_board(board)
    print("Winner:", winner(board))

play()
```

```python
#BTTL1
import copy
import math
import random
import numpy as np

# Khởi tạo các hằng số đại diện cho người chơi và trạng thái ô trống
X = "X"
O = "O"
EMPTY = None
user = None
ai = None

# CÁC HÀM QUẢN LÝ TRẠNG THÁI
def initial_state():
    """Returns starting state of the board."""
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    """Returns player who has the next turn on a board."""
    count = 0
    for i in board:
        for j in i:
            if j: # Nếu ô đã có X hoặc O
                count += 1
    # Nếu số nước đi là lẻ, đến lượt AI, ngược lại là User
    if count % 2 != 0:
        return ai
    return user

def actions(board):
    """Returns set of all possible actions (i, j) available on the board."""
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    """Returns the board that results from making move (i, j) on the board."""
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board

# CÁC HÀM KIỂM TRA THẮNG/THUA
def get_horizontal_winner(board):
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[i][0]
        for j in range(board_len):
            if board[i][j] != winner_val:
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val

def get_vertical_winner(board):
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[0][i]
        for j in range(board_len):
            if board[j][i] != winner_val:
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val

def get_diagonal_winner(board):
    winner_val = None
    board_len = len(board)
    # Kiểm tra đường chéo chính
    winner_val = board[0][0]
    for i in range(board_len):
        if board[i][i] != winner_val:
            winner_val = None
    if winner_val:
        return winner_val

    # Kiểm tra đường chéo phụ
    winner_val = board[0][board_len - 1]
    for i in range(board_len):
        j = board_len - 1 - i
        if board[i][j] != winner_val:
            winner_val = None
    return winner_val

def winner(board):
    """Returns the winner of the game, if there is one."""
    winner_val = get_horizontal_winner(board) or get_vertical_winner(board) or get_diagonal_winner(board) or None
    return winner_val

def terminal(board):
    """Returns True if game is over, False otherwise."""
    if winner(board) != None:
        return True
    for i in board:
        for j in i:
            if j == EMPTY:
                return False
    return True

def utility(board):
    """Returns 1 if X has won the game, -1 if O has won, 0 otherwise."""
    winner_val = winner(board)
    if winner_val == X:
        return 1
    elif winner_val == O:
        return -1
    return 0
# THUẬT TOÁN MINIMAX
def maxValue(state):
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v

def minValue(state):
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v

def minimax(board):
    """Returns the optimal action for the current player on the board."""
    current_player = player(board)
    move = None

    if current_player == X:
        best_val = -math.inf
        for action in actions(board):
            check = minValue(result(board, action))
            if check > best_val:
                best_val = check
                move = action
    else:
        best_val = math.inf
        for action in actions(board):
            check = maxValue(result(board, action))
            if check < best_val:
                best_val = check
                move = action
    return move
board = initial_state()
ai_turn = False

print("Choose a player (X or O):")
user = input().strip().upper()
if user == "X":
    ai = "O"
else:
    ai = "X"

print("\nGame Start!")
print(np.array(board))

while True:
    game_over = terminal(board)
    playr = player(board)

    if game_over:
        win = winner(board)
        print("\n====================")
        if win is None:
            print("Game Over: Tie.")
        else:
            print(f"Game Over: {win} wins.")
        print("====================")
        break
    else:
        # Lượt của AI
        if user != playr and not game_over:
            if ai_turn:
                print(f"\n--- AI ({ai}) is thinking... ---")
                move = minimax(board)
                board = result(board, move)
                ai_turn = False
                print("AI Move:")
                print(np.array(board))

        # Lượt của User
        elif user == playr and not game_over:
            ai_turn = True
            print(f"\n--- Your turn ({user}) ---")
            print("Enter the position to move (row,col) from 0 to 2")
            try:
                i = int(input("Row (0, 1, 2): "))
                j = int(input("Col (0, 1, 2): "))
                if board[i][j] == EMPTY:
                    board = result(board, (i, j))
                    print("Your Move:")
                    print(np.array(board))
                else:
                    print("Invalid move. Cell is not empty! Try again.")
                    ai_turn = False # Trả lại lượt cho user
            except (ValueError, IndexError):
                print("Invalid input. Please enter numbers between 0 and 2.")
                ai_turn = False # Trả lại lượt cho user
```

```python
#BTTL2
import math
from IPython.display import clear_output

def GetWinner(board):
    """Kiểm tra người chiến thắng, trả về 'X', 'O' hoặc None"""
    # Kiểm tra hàng ngang
    if board[0] == board[1] and board[1] == board[2]: return board[0]
    elif board[3] == board[4] and board[4] == board[5]: return board[3]
    elif board[6] == board[7] and board[7] == board[8]: return board[6]

    # Kiểm tra hàng dọc
    elif board[0] == board[3] and board[3] == board[6]: return board[0]
    elif board[1] == board[4] and board[4] == board[7]: return board[1]
    elif board[2] == board[5] and board[5] == board[8]: return board[2]

    # Kiểm tra đường chéo
    elif board[0] == board[4] and board[4] == board[8]: return board[0]
    elif board[2] == board[4] and board[4] == board[6]: return board[2]

    return None

def PrintBoard(board):
    """Xóa màn hình và in bảng Tic-Tac-Toe hiện tại"""
    clear_output(wait=True)
    print(f'''
     {board[0]} | {board[1]} | {board[2]}
    ---|---|---
     {board[3]} | {board[4]} | {board[5]}
    ---|---|---
     {board[6]} | {board[7]} | {board[8]}
    ''')

def GetAvailableCells(board):
    """Lấy danh sách các ô còn trống"""
    available = []
    for cell in board:
        if cell != "X" and cell != "O":
            available.append(cell)
    return available

# THUẬT TOÁN ALPHA-BETA PRUNING
def minimax(position, depth, alpha, beta, isMaximizing):
    """Thuật toán Minimax kết hợp cắt tỉa Alpha-Beta"""
    winner = GetWinner(position)

    # Hàm đánh giá (Utility function) tính thêm độ sâu để AI ưu tiên thắng nhanh / thua chậm
    if winner != None:
        return 10 - depth if winner == "X" else -10 + depth

    if len(GetAvailableCells(position)) == 0:
        return 0

    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "X"
            Eval = minimax(position, depth + 1, alpha, beta, False)
            maxEval = max(maxEval, Eval)
            alpha = max(alpha, Eval)
            position[cell - 1] = cell # Trả lại trạng thái cũ (Backtrack)

            if beta <= alpha:
                break  # Prune (Cắt tỉa nhánh)
        return maxEval
    else:
        minEval = math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "O"
            Eval = minimax(position, depth + 1, alpha, beta, True)
            minEval = min(minEval, Eval)
            beta = min(beta, Eval)
            position[cell - 1] = cell # Trả lại trạng thái cũ (Backtrack)

            if beta <= alpha:
                break  # Prune (Cắt tỉa nhánh)
        return minEval

def FindBestMove(currentPosition, AI):
    """Tìm nước đi tối ưu dựa trên điểm số đánh giá từ hàm minimax"""
    bestVal = -math.inf if AI == "X" else math.inf
    bestMove = -1

    for cell in GetAvailableCells(currentPosition):
        currentPosition[cell - 1] = AI
        # AI "X" là MAX player, "O" là MIN player
        moveVal = minimax(currentPosition, 0, -math.inf, math.inf, False if AI == "X" else True)
        currentPosition[cell - 1] = cell # Backtrack

        if (AI == "X" and moveVal > bestVal):
            bestMove = cell
            bestVal = moveVal
        elif (AI == "O" and moveVal < bestVal):
            bestMove = cell
            bestVal = moveVal

    return bestMove

# CHƯƠNG TRÌNH CHÍNH
def main():
    print("Play as X or O? (X goes first)")
    player = input().strip().upper()
    AI = "O" if player == "X" else "X"

    # Khởi tạo bảng với các số từ 1 đến 9
    currentGame = [*range(1, 10)]
    currentTurn = "X"  # X luôn đi trước

    while True:
        # TÌM NGƯỜI CHIẾN THẮNG TRƯỚC KHI CHUYỂN LƯỢT
        win = GetWinner(currentGame)
        if win != None:
            PrintBoard(currentGame)
            print("====================")
            print(f"Game Over: {win} WON!!!")
            print("====================")
            break

        if len(GetAvailableCells(currentGame)) == 0:
            PrintBoard(currentGame)
            print("====================")
            print("Game Over: Tie.")
            print("====================")
            break

        # ĐẾN LƯỢT AI
        if currentTurn == AI:
            print(f"\nAI ({AI}) is thinking...")
            cell = FindBestMove(currentGame, AI)
            currentGame[cell - 1] = AI
            currentTurn = player

        # ĐẾN LƯỢT NGƯỜI CHƠI
        elif currentTurn == player:
            PrintBoard(currentGame)
            while True:
                try:
                    humanInput = int(input(f"Your turn ({player}). Enter Number (1-9): ").strip())
                    if humanInput in currentGame:
                        currentGame[humanInput - 1] = player
                        currentTurn = AI
                        break
                    else:
                        print("Invalid move. Cell is not available! Try again.")
                except ValueError:
                    print("Invalid input. Please enter numbers between 1 and 9.")

if __name__ == "__main__":
    main()
```
