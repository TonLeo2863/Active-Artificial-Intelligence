```python
#Bai13
def tron_mang(a, b):
    kq = []
    min_len = min(len(a), len(b))
    for i in range(min_len):
        kq.append(a[i] + b[i])
    if len(a) > min_len:
        kq.extend(a[min_len:])
    if len(b) > min_len:
        kq.extend(b[min_len:])
    return kq
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print(tron_mang(a, b))
```

```python
#Bai14
import random

def tao_mang(m, n):
    return [[random.randint(1, 50) for _ in range(n)] for _ in range(m)]

def xuat_dong(a, k):
    return a[k]

def xuat_cot(a, k):
    return [row[k] for row in a]

def dong_tong_lon_nhat(a):
    tong_dong = [__builtins__.sum(row) for row in a]
    return max(range(len(tong_dong)), key=lambda i: tong_dong[i])

def cot_tich_nho_nhat(a):
    m, n = len(a), len(a[0])
    tich_cot = []
    for j in range(n):
        tich = 1
        for i in range(m):
            tich *= a[i][j]
        tich_cot.append(tich)
    return min(range(n), key=lambda j: tich_cot[j])

def xuat_dong_chan_cot_le(a):
    m, n = len(a), len(a[0])
    return [a[i][j] for i in range(m) if i % 2 == 0 for j in range(n) if j % 2 == 1]

def tb_chan_dong_le(a):
    m, n = len(a), len(a[0])
    ds = [a[i][j] for i in range(m) if i % 2 == 1 for j in range(n) if a[i][j] % 2 == 0]
    return __builtins__.sum(ds)/len(ds) if ds else None

def tb_bien(a):
    m, n = len(a), len(a[0])
    bien = []
    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                bien.append(a[i][j])
    return __builtins__.sum(bien)/len(bien)

def tb_tich_khong_bien(a):
    m, n = len(a), len(a[0])
    ds = []
    for i in range(1, m-1):
        for j in range(1, n-1):
            ds.append(a[i][j])
    tich = 1
    for x in ds:
        tich *= x
    return tich/len(ds) if ds else None

m, n = 3, 4
my_matrix = tao_mang(m, n)
print("Ma trận ngẫu nhiên:")
for row in my_matrix:
    print(row)

print(f"Dòng thứ 1: {xuat_dong(my_matrix, 0)}")
print(f"Cột thứ 2: {xuat_cot(my_matrix, 1)}")
print(f"Dòng có tổng lớn nhất (chỉ số): {dong_tong_lon_nhat(my_matrix)}")
print(f"Cột có tích nhỏ nhất (chỉ số): {cot_tich_nho_nhat(my_matrix)}")
print(f"Phần tử dòng chẵn cột lẻ: {xuat_dong_chan_cot_le(my_matrix)}")
print(f"Trung bình cộng các số chẵn trên dòng lẻ: {tb_chan_dong_le(my_matrix)}")
print(f"Trung bình cộng các phần tử biên: {tb_bien(my_matrix)}")
print(f"Trung bình nhân các phần tử không biên: {tb_tich_khong_bien(my_matrix)}")
```

```python
#Bai15
class SinhVien:
    def __init__(self, ma, ten, nam_sinh, diem_tb):
        self.ma = ma
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb

def nhap_ds_sv():
    ds = []
    n = int(input("Nhập số lượng sinh viên: "))
    for i in range(n):
        print(f"\nSinh viên thứ {i+1}")
        ma = input("Mã sinh viên (10 ký tự): ")
        ten = input("Tên sinh viên: ")
        nam_sinh = int(input("Năm sinh: "))
        diem_tb = float(input("Điểm trung bình: "))
        ds.append(SinhVien(ma, ten, nam_sinh, diem_tb))
    return ds

def sv_du_dk_len_lop(ds):
    return [sv for sv in ds if sv.diem_tb >= 5]

def sv_du_20_tuoi(ds, nam_hien_tai):
    return [sv for sv in ds if nam_hien_tai - sv.nam_sinh == 20]

def dem_sv_he_DH(ds):
    return __builtins__.sum(1 for sv in ds if sv.ma[2:4] == "DH")

def dem_sv_ten_Lan(ds):
    return __builtins__.sum(1 for sv in ds if sv.ten.strip() == "Lan")

def dem_sv_ho_Phan(ds):
    return __builtins__.sum(1 for sv in ds if sv.ten.split()[0] == "Phan")

if __name__ == "__main__":
    ds = nhap_ds_sv()
    nam_hien_tai = 2026

    print("\sinh viên đủ điều kiện lên lớp:")
    for sv in sv_du_dk_len_lop(ds):
        print("-", sv.ten)

    print("Sinh viên đủ 20 tuổi:")
    for sv in sv_du_20_tuoi(ds, nam_hien_tai):
        print("-", sv.ten)

    print("Số sinh viên hệ DH:", dem_sv_he_DH(ds))
    print("Số sinh viên tên Lan:", dem_sv_ten_Lan(ds))
    print("Số sinh viên họ Phan:", dem_sv_ho_Phan(ds))
```

```python
#Bài về nhà bài 1
def tron_mang_min(a, b):
    n = min(len(a), len(b))
    res = [min(a[i], b[i]) for i in range(n)]
    res += a[n:] if len(a) > n else b[n:]
    return res
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print(f"Kết quả trộn: {tron_mang_min(a, b)}")
```
