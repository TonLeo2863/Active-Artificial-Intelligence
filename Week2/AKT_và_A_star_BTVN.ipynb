```python
#BTVN1
import heapq
import copy
class Node:
    def __init__(self, state, parent=None, move=None, g=0):
        self.state = state
        self.parent = parent
        self.move = move
        self.g = g
        self.h = self.heuristic()
        self.f = self.g + self.h

    def heuristic(self):
        """Manhattan Distance"""
        n = len(self.state)
        dist = 0
        for i in range(n):
            for j in range(n):
                val = self.state[i][j]
                if val != 0:
                    target_x = (val - 1) // n
                    target_y = (val - 1) % n
                    dist += abs(i - target_x) + abs(j - target_y)
        return dist

    def __lt__(self, other):
        return self.f < other.f
def find_zero(state):
    n = len(state)
    for i in range(n):
        for j in range(n):
            if state[i][j] == 0:
                return i, j


def get_neighbors(state):
    n = len(state)
    x, y = find_zero(state)
    moves = []
    directions = [(-1,0),(1,0),(0,-1),(0,1)]

    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if 0 <= nx < n and 0 <= ny < n:
            new_state = copy.deepcopy(state)
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            moves.append(new_state)
    return moves


def is_goal(state):
    n = len(state)
    goal = list(range(1, n*n)) + [0]
    flat = sum(state, [])
    return flat == goal


def reconstruct_path(node):
    path = []
    while node:
        path.append(node.state)
        node = node.parent
    return path[::-1]
def astar(start):
    open_list = []
    closed_set = set()

    start_node = Node(start)
    heapq.heappush(open_list, start_node)

    while open_list:
        current = heapq.heappop(open_list)

        if is_goal(current.state):
            return reconstruct_path(current)

        closed_set.add(str(current.state))

        for neighbor in get_neighbors(current.state):
            if str(neighbor) in closed_set:
                continue

            child = Node(neighbor, current, g=current.g + 1)
            heapq.heappush(open_list, child)

    return None
start_state = [
    [1, 2, 3, 4],
    [5, 6, 0, 8],
    [9,10,7,11],
    [13,14,15,12]
]

solution = astar(start_state)

if solution:
    print("Số bước:", len(solution)-1)
    for step in solution:
        for row in step:
            print(row)
        print("------")
else:
    print("Không tìm được lời giải")
```

```python
#BTVN2
import heapq

class TSP_AStar:
    def __init__(self, matrix):
        self.matrix = matrix
        self.n = len(matrix)

    def h(self, current_city, unvisited):
        if not unvisited:
            return self.matrix[current_city][0]
        return min([self.matrix[current_city][city] for city in unvisited])

    def solve(self):
        start_node = (0, 0, 0, 1 << 0)
        open_lst = [start_node]

        while open_lst:
            f, g, u, mask = heapq.heappop(open_lst)
            if mask == (1 << self.n) - 1:
                return g + self.matrix[u][0]
            for v in range(self.n):
                if not (mask & (1 << v)):
                    new_g = g + self.matrix[u][v]
                    new_mask = mask | (1 << v)
                    new_f = new_g + self.h(v, [i for i in range(self.n) if not (new_mask & (1 << i))])
                    heapq.heappush(open_lst, (new_f, new_g, v, new_mask))
        return None
dist_matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

tsp = TSP_AStar(dist_matrix)
print(f"Chi phí đường đi ngắn nhất (TSP): {tsp.solve()}")
```
